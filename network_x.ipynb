```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Neusmjereni graf
G = nx.Graph()

# dodaj vrhove
#G.add_node(1)
G.add_nodes_from([2,4])

# dodaj bridove
#G.add_edge(1,2)
# slozimo primjer s predavanja
G.add_edges_from([(1,2),(1,5),(2,3),(2,4),(2,5),(3,4),(4,5)])

# popis vrhova
print(G.nodes())
# popis bridova
print(G.edges())

# ilustracija grafa
nx.draw(G)
# stupnjevi svih vrhova u dictionaryu
nx.degree(G)

nx.find_cycle(G)
```

```python
GC = nx.complement(G)
nx.draw(GC)
```

```python
# Usmjereni graf
DG = nx.DiGraph()
DG.add_weighted_edges_from([(1,2,0.5),(3,1,0.75)])  #s te≈æinama
nx.draw(DG)
```

```python
DGT = nx.reverse(DG) #transpose
nx.draw(DGT)
```

```python
# imenovani grafovi
petersen = nx.petersen_graph()
#nx.draw_circular(petersen)

K_5 = nx.complete_graph(5)
nx.draw(K_5)
```

```python
# dodavanje oznaka vrhovima
G = nx.Graph()

#G.add_edges_from([(1,2),(1,3)])
G.add_edges_from([(1,2),(1,5),(2,3),(2,4),(2,5),(3,4),(4,5)])
G.remove_edges_from([(1,5),(1,2)])

# pozicioniranje
pos = nx.spring_layout(G)

# vizualizacija vrhova
#alpha je prozirnost
nx.draw_networkx_nodes(
    G,
    pos,
    nodelist = G.nodes(),
    node_color = 'red',
    node_size = 700,
    alpha = 0.8
)

# vizualizacija bridova
nx.draw_networkx_edges(G,pos,width=1.0,alpha=0.5)

# vizualizacija oznaka
labels = {1:r'$1$',2:r'$2$',3:r'$3$',4:r'$4$',5:r'$5$'}  # dolar zbog latexa npr.$x^2$

nx.draw_networkx_labels(G,pos,labels,font_size=16)

plt.axis('off')
plt.show()


# je li povezan?
print(nx.is_connected(G))
print(nx.number_connected_components(G))
```
